# Álgebra Lineal con Python
Este cuaderno muestra conceptos básicos de álgebra lineal aplicados con librerías de Python:
- **NumPy** para operaciones con matrices y vectores
- **Pandas** para manipulación de datos tabulares
- **Matplotlib** para visualización
- **Scikit-learn** para aplicaciones en Machine Learning

# Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## Operaciones con vectores y matrices
Ejemplos de suma, resta, producto punto y transposición.

```python
# Definición de vectores y matrices
A = np.array([[1, 2], [3, 4]])
B = np.array([[2, 0], [1, 3]])
v = np.array([1, 2])

# Operaciones
print("Suma:\n", A + B)
print("Resta:\n", A - B)
print("Multiplicación:\n", A @ B)
print("Transpuesta:\n", A.T)
print("Inversa:\n", np.linalg.inv(A))
print("Producto punto vectorial:", np.dot(v, v))
```

## Pandas y álgebra lineal
Crearemos un DataFrame y aplicaremos operaciones sobre columnas.

```python
df = pd.DataFrame({
    "x": [1, 2, 3, 4, 5],
    "y": [2, 4, 6, 8, 10]
})

print(df)

# Producto escalar entre columnas
producto = np.dot(df["x"], df["y"])
print("Producto escalar entre columnas x e y:", producto)
```

## Visualización de vectores y transformaciones
Graficaremos vectores en 2D y una transformación lineal.

```python
# Graficar vectores
plt.quiver(0,0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r')
plt.xlim(-1,3)
plt.ylim(-1,3)
plt.grid()
plt.show()

# Transformación lineal
trans_v = A @ v
plt.quiver(0,0, trans_v[0], trans_v[1], angles='xy', scale_units='xy', scale=1, color='b')
plt.xlim(-1,6)
plt.ylim(-1,6)
plt.grid()
plt.show()
```

## Regresión lineal
Ejemplo práctico de cómo el álgebra lineal se aplica en Machine Learning.

```python
X = df[["x"]]
y = df["y"]

modelo = LinearRegression()
modelo.fit(X, y)

print("Coeficiente:", modelo.coef_)
print("Intercepto:", modelo.intercept_)

# Predicciones
y_pred = modelo.predict(X)

# Visualización
plt.scatter(X, y, color="blue")
plt.plot(X, y_pred, color="red")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Regresión lineal con Scikit-learn")
plt.show()
```
